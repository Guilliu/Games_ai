<span style='color:blue'>Reseteamos el juego y creamos dos agentes (uno de cada color) que juegen una partida con 2 movimientos cada uno

```python
movements_each = 2

obs = game_obs()
obs.place_pieces()
obs.compute_legal_moves()

agente1 = agent('b', info=True)
agente2 = agent('r', info=True)

for i in range(2*movements_each):
    if i%2 == 0: agente1.play(obs)
    else: agente2.play(obs)
obs.display_board()
```

<span style='color:blue'>Probamos un nuevo agente que elige (si puede) una pieza aleatoria entre las que pueden capturar y realiza uno de los movimientos que acaba en captura. Generamos una situación en la que la dama azul de (3, 7) puede capturar cualquiera de los tres peones situados en: (3, 4), (2,6), (1,7)

```python
obs = game_obs()
obs.place_pieces()
obs.compute_legal_moves()

obs.move_piece((6, 4), (4, 4))
obs.move_piece((1, 4), (3, 4))
obs.move_piece((7, 3), (3, 7))
obs.move_piece((1, 6), (2, 6))
obs.display_board()

agente = agent('b', stgy_selection='pro1', stgy_move='pro1', info=True)
agente.play(obs)
obs.display_board()
```
